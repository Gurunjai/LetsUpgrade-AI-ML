```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('datasets_13720_18513_insurance.csv')
dataset.head()
```

```python
dataset.shape
```

```python
sns.lmplot(x='bmi', y='charges', data=dataset, aspect=2, height=6)
plt.xlabel("Body Mass Index $(kg/m^2)$: as Independent variable")
plt.ylabel("Insurance Charges: as Dependent variable")
plt.title("Charge Vs BMI")
```

```python
sns.lmplot(x='age', y='charges', data=dataset, aspect=2, height=6)
plt.xlabel("Age as independent variable")
plt.ylabel("Insurance Charges: as Dependent variable")
plt.title("Charge Vs Age")
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12, 4))
sns.heatmap(dataset.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing value in the dataset')
```

```python
#correlation plot
corr = dataset.corr()
sns.heatmap(corr, cmap='Wistia', annot=True);
```

```python
f = plt.figure(figsize=(12, 4))

ax = f.add_subplot(121)
sns.distplot(dataset['charges'], bins=50, color='r', ax=ax)
ax.set_title('Distribution of Insurance charges')

ax = f.add_subplot(122)
sns.distplot(np.log10(dataset['charges']), bins=40, color='b', ax=ax)
ax.set_title('Distribution of Insurance charges in $log$ scale')
ax.set_xscale('log');
```

```python
f = plt.figure(figsize=(14, 6))
ax = f.add_subplot(121)
sns.violinplot(x='sex', y='charges', data=dataset, palette='Wistia', ax=ax)
ax.set_title('Violin plot of charges vs sex')

ax = f.add_subplot(122)
sns.violinplot(x='smoker', y='charges', data=dataset, palette='magma', ax=ax)
ax.set_title('Violin plot of charges vs smoker');
```

```python
f = plt.figure(figsize=(14, 6))
sns.boxplot(x='children', y='charges', hue='sex', data=dataset, palette='rainbow')
plt.title('Box plot of charges vs childern');
```
